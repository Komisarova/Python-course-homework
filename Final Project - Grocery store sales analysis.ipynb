За допомогою цих даних ми спостерігаємо за динамікою споживчого попиту -про те, як змінюється обсяг продажів різних категорій товарів протягоомВ даному випадку, бачимо, що в абсолютно усіх категоріях позитивна динаміка протягом 2016-2019 років. Є з більшою часткою, є з меншою, але негативний тренд відсутній по товарам. Далі, візуалізуємо дані за весь час роботи магазинів для розуміння частки категорій товарів у загальному обсягу. Це нам дасть розуміння, який асортимент товарів краще розширювати живчого попиту

```python
import matplotlib.pyplot as plt
```

```python
pivot_table = df.pivot_table(values='Gross Bill Amount', index='Year', columns='Category', aggfunc='sum')
total_sales_by_category = pivot_table.sum()
plt.figure(figsize=(10, 8))
plt.pie(total_sales_by_category, labels=total_sales_by_category.index, autopct='%1.1f%%', startangle=140)
plt.title('Загальний обсяг продажів за категоріями товарів за 2016-2019 роки')
plt.show()
```

```python
#Розглянемо як змінюється обсяг продажів по місяцях
sales_by_month_year = df.groupby(['Year', 'Month'])['Gross Bill Amount'].sum().unstack()
```

```python
[plt.plot(sales_by_month_year.columns, sales_by_month_year.loc[year], marker='o', label=year) for year in sales_by_month_year.index]
plt.title('Динаміка обсягу продажів')
plt.xlabel('Місяць')
plt.ylabel('Обсяг продажів')
plt.xticks(range(1, 13), ['Січ', 'Лют', 'Бер', 'Кві', 'Тра', 'Чер', 'Лип', 'Сер', 'Вер', 'Жов', 'Лис', 'Гру'])
plt.legend(title='Рік')
plt.grid(True)
```

Висновок - протягом років пік продажу припадає на грудень. ПОряд з цим, у 2018 році дещо інший тренд і у липні обсяг продажу був більше ніж у грудні. Для більш детального аналізу можна порівняти категорії товарів (можливо було постачання нової продукції/позиції або акції на якийсь товар, що призвело до більшого попиту). Також на майбутнє можна розглянути виокремити місяці з низьким обсягом продажу для залучення покупців - акції, дегустації та ін.

```python
df.groupby('City')['Profit Margin'].sum().plot(kind='bar', color='skyblue')
plt.title('Загальний прибуток за містами')
plt.xlabel('Місто')
plt.ylabel('Прибуток')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
```

Найбільш прибуткові магазини у Bedok та Manila

5. Кореляція між віком клієнтів та сумою їхніх покупок

```python
correlation = df['Age'].corr(df['Gross Bill Amount'])

print("Коефіцієнт кореляції між віком клієнтів та сумою їхніх покупок:", correlation)
```

Застосовано коефіцієнт кореляції Пірсона (ступінь лінійної залежності між двома змінними - чим ближче значення коефіцієнта до 1 або -1, тим сильніше кореляція) Отримане значення дуже близьке до нуля (0.006), що означає дуже слабку залежність між цими двома змінними. З цього можна зробити висновок, що вік клієнтів не має значущого впливу на суму їхніх покупок в даному наборі даних.

```python
df.groupby('Country')['Mobile Number'].nunique()
```

```python
#тепер візуально поглянемо на розподіл
plt.figure(figsize=(8, 8))
plt.pie(df.groupby('Country')['Mobile Number'].nunique(), labels=df['Country'].unique(), autopct='%1.1f%%', startangle=140)
plt.title('Доля покупців у розрізі країн')
plt.axis('equal')
plt.show()
```

Половина покупців мережі здійснює покупки у Сінгапурі. Можна розглянути питання залучення реклами в країнах з меншою долею покупців та відкриття додаткових магазинів у Сінгапурі.
